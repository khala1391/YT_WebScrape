## import library

```python
from urllib.request import urlopen # option#1
import requests  # option#2

import pandas as pd
from bs4 import BeautifulSoup

import time
```

```python
url = "https://www.naiin.com/category?category_1_code=13&product_type_id=1"  # for long run test (greater than 50 pages)
# url = "https://www.naiin.com/category?magazineHeadCode=BS&product_type_id=2"   # for error test (only 5 pages available)
r = requests.get(url)
print(r)
print(r.url)
soup = BeautifulSoup(r.text, "lxml")
```

### for loop
- for validity test

```python
books = soup.find_all("a", class_="itemname")
book_list = [book.text.strip() for book in books]
print(book_list)
print(len(book_list))

authors = soup.find_all("a", class_="inline-block tw-whitespace-normal tw-block")
author_list = [author.text.strip() for author in authors]
print(author_list)
print(len(author_list))

prices = soup.find_all("p", class_="txt-price")
price_list = [price.text.strip() for price in prices]
print(price_list)
print(len(price_list))
```

```python
i = 0
n = 3

for i in range(n):
    try:
        np = soup.find("a", class_= "nav-pag pag-next").get("href")
        print(np)

        url = np
        r = requests.get(url)
        soup = BeautifulSoup(r.text, "lxml")
        time.sleep(5)
        
        books = soup.find_all("a", class_="itemname")      
        book_list.extend([book.text.strip() for book in books])
        print(book_list[::-1])
        print(len(book_list))
        
        authors = soup.find_all("a", class_="inline-block tw-whitespace-normal tw-block")
        author_list.extend([author.text.strip() for author in authors])
        print(author_list[::-1])
        print(len(author_list))
        
        prices = soup.find_all("p", class_="txt-price")
        price_list.extend([price.text.strip() for price in prices])
        print(price_list[::-1])
        print(len(price_list))
        
        i += 1
        
    except Exception as e:
        print(f"Error: {e}")
        break    
```

### while loop
- for full web scrape

```python
books = soup.find_all("a", class_="itemname")
book_list = [book.text.strip() for book in books]
print(book_list)
print(len(book_list))


authors = soup.find_all("a", class_="inline-block tw-whitespace-normal tw-block")
author_list = [author.text.strip() for author in authors]
print(author_list)
print(len(author_list))

prices = soup.find_all("p", class_="txt-price")
price_list = [price.text.strip() for price in prices]
print(price_list)
print(len(price_list))
```

```python
while True:
    try:
        np = soup.find("a", class_= "nav-pag pag-next").get("href")
        print(np)

        url = np
        r = requests.get(url)
        soup = BeautifulSoup(r.text, "lxml")
        time.sleep(2)
        
        books = soup.find_all("a", class_="itemname")      
        book_list.extend([book.text.strip() for book in books])
        # print(book_list[::-1])
        print(len(book_list))

        authors = soup.find_all("a", class_="inline-block tw-whitespace-normal tw-block")
        author_list.extend([author.text.strip() for author in authors])
        # print(author_list[::-1])
        print(len(author_list))
        
        prices = soup.find_all("p", class_="txt-price")
        price_list.extend([price.text.strip() for price in prices])
        # print(price_list[::-1])
        print(len(price_list))
        
        
        
    except Exception as e:
        print(f"Error: {e}")
        print(len(book_list))
        print(len(author_list))
        print(len(price_list))
        break
```

```python
df = pd.DataFrame({"book": book_list,
                   "author": author_list,
                  "price": price_list,})
df.to_csv('data/book_details.csv',encoding='utf-8-sig')
```
