## import library

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url="https://www.soccersuck.com/"
r = requests.get(url)
# print(r)
r.status_code
```

## set up BeautifulSoup

```python
soup = BeautifulSoup(r.text, "lxml")
table = soup.find("div", class_= "lastpanel1")
print(table)
```

## collect data to list

```python
hl = table.find_all('a')
# hl = [i.text for i in hl]
hl_text = [i.get_text(strip=True) for i in hl]
print(hl_text)
len(hl_text)

# float(i.get_text(strip=True))
```

```python
hl_link = [i.get('href') for i in hl]
print(hl_link)
len(hl_link)
```

```python
soup = BeautifulSoup(r.text, "lxml")
table2 = soup.find("div", class_= "lastpanel2")
# print(table2)

hl2 = table2.find_all('a')
# hl = [i.text for i in hl]
hl2_text = [i.get_text(strip=True) for i in hl2]
print(hl2_text)
# len(hl2_text)

hl2_link = [i.get('href') for i in hl2]
print(hl2_link)
display(len(hl2_link))
```

## save data to dataframe

```python
hl_text_combined = hl_text + hl2_text
hl_link_combined = hl_link + hl2_link
df = pd.DataFrame({'text':hl_text_combined,'link':hl_link_combined})
df

```
